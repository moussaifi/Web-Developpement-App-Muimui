```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
import random

url = "https://www.macys.com/shop/womens-clothing/all-womens-clothing?id=188851&edge=hybrid"
```

```python
main_page = requests.get(url, headers={'User-Agent' : "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/33.0.1750.149 Safari/537.36"}).content
main_soup = BeautifulSoup(main_page, "lxml")
```

```python
prod_pages = []
prod_names = []
prod_prices = []
prod_dp = []
prod_images = []
prod_ratings = []
prod_brands = []
delay=(1,3)
```

```python
pages=[url]
page = main_page
soup = main_soup
while True:
    for item in soup.find_all('a', {'class': 'productDescLink'}):
        href = item.attrs['href']
        item_link = "https://www.macys.com/"+href
        prod_pages.append(item_link)
        
        item_page = requests.get(item_link, headers={'User-Agent' : "Mozilla/5.0"}).content
        item_soup = BeautifulSoup(item_page, "lxml")

        name = item_soup.find('h1', {'data-auto':"product-name"}).text.strip('\n ')
        prod_names.append(name)

        price = item_soup.find('div', {'class': 'price'}).text
        price = float(price[2:])
        prod_prices.append(price)
        
        brand = item_soup.find('a', {'class': "c-reverse-link"}).text.strip('\n    ')
        prod_brands.append(brand)
        

        if item_soup.find('span', {'class': 'red-star'}) is not None:
            rating = item_soup.find('span', {'class': 'red-star'}).attrs['style'].strip("width: %")
            prod_ratings.append((float(rating)*5)/100)
        else: prod_ratings.append(0)
            

        dp1 = item_soup.find('p', {'data-auto':"product-description"}).text.strip('\n ')
        dp2 = item_soup.find('ul', {'class': 'reset-font'}).text.strip('\n ').split('\n')
        dp2 = " ".join(dp2)
        prod_dp.append(dp1+" "+dp2)
        

        image = item_soup.find('source', {'type': "image/webp"}).attrs['srcset']
        prod_images.append(image) 
    next_page = soup.find('li',{'class':"next-page"})
    if next_page is None: break
    next_page_link = "https://www.macys.com"+next_page.div.a.attrs['href']
    pages.append(next_page_link)
    time.sleep(random.randint(delay[0],delay[1]))
    page = requests.get(next_page_link, headers={'User-Agent' : "Mozilla/5.0"}).content
    soup = BeautifulSoup(page, "lxml")
```

```python
d = {'prod_page_link': prod_pages, 'image_link': prod_images,
        'name': prod_names, 'price': prod_prices, 'description': prod_dp, 'brand': prod_brands, 'rating': prod_ratings}
df = pd.DataFrame(d)
```
